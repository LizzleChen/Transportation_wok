```python
import numpy as np
import pandas as pd
```

```python
s4=pd.read_csv('S4_I-35E.csv')
s4_tmc=pd.read_csv('TMC_Identification.csv')
```

```python
s4.head(5)
```

```python
s4_tmc.head(10)
```

```python
df=pd.merge(s4,s4_tmc,left_on='tmc_code',right_on='tmc',how='left')
```

```python
df.columns
```

```python
df1=df[['tmc_code','measurement_tstamp','speed','average_speed','travel_time_seconds', 'direction','miles']]
```

```python
print(df1)
```

```python
import datetime
```

```python
df1.loc[:,'datetime']=pd.to_datetime(df1['measurement_tstamp'])
```

```python
df1['datetime'].dtype
```

```python
df1.head(5)
```

```python
df1.loc[:,'date']=df1['datetime'].dt.date
df1.loc[:,'time']=df1['datetime'].dt.time
df1.drop(['measurement_tstamp'],axis=1)
```

```python
firstdate=pd.to_datetime("2018-06-01").date()
seconddate=pd.to_datetime("2018-08-31").date()
df1=df1[(df1['date']<firstdate) | (df1['date']>seconddate)]
```

```python
df1=df1[df1['time'] < pd.to_datetime("18:30:00").time()]
```

```python
df1.head
```

```python
df1['weekday']=df1['datetime'].dt.dayofweek
```

```python
df1 = df1.dropna(how='any',axis=0) 
```

```python
df1.loc[df1['time']<pd.to_datetime('9:00:00').time(),'peak']='AM'
df1.loc[df1['time']>=pd.to_datetime('15:00:00').time(),'peak']='PM'
df1.loc[:,'peak']=df1['peak'].fillna('OP')
```

```python
df1
```

```python
tot_avg=df1.pivot_table(index=['direction','tmc_code'],values='travel_time_seconds',aggfunc='sum',columns=['peak'])
```

```python
tot_avg
```

```python
tot_counts=df1.pivot_table(index=['direction','tmc_code'],values='travel_time_seconds',aggfunc='count',columns=['peak'])
tot_counts
```

```python
avg=tot_avg.div(tot_counts)
avg
```
